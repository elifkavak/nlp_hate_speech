```python
def train_naive_bayes(freqs, train_x, train_y):
    loglikelihood = {}
    logprior = 0
    
    # calculate the number of unique words in vocabulary
    unique_words = set([pair[0] for pair in freqs.keys()])
    V = len(unique_words)
    
    # calculate N_pos and N_neg
    N_pos = N_neg = 0
    for pair in freqs.keys():
        if pair[1] > 0:
            N_pos += freqs[(pair)]
        else:
            N_neg += freqs[(pair)]
        
    # calc the number of documents (tweets)
    D = train_y.shape[0]
    
    # calc D_pos, the number of positive documents (tweets)
    D_pos = sum(train_y)
    
    # calc D_neg, the number of negative documents (tweets)
    D_neg = D - sum(train_y)
    
    # calc logprior
    logprior = np.log(D_pos) - np.log(D_neg)
    
    # for each unique word
    for word in unique_words:
        
        # get the positive and negative freq of the word
        freq_pos = freqs.get((word, 1), 0)
        freq_neg = freqs.get((word, 0), 0)
        
        # calculate the probability that word is positive, and negative
        p_w_pos = (freq_pos + 1) / (N_pos + V)
        p_w_neg = (freq_neg + 1) / (N_pos + V)
        
        # calc the log likelihood of the word
        loglikelihood[word] = np.log(p_w_pos / p_w_neg)
        
    return logprior, loglikelihood
```
